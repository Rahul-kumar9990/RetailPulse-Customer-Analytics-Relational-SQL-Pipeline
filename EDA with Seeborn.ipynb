# 1. Which category generates the most revenue?

```python
# Install Seaborn (if not already installed)
!pip install seaborn

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the csv file
df_merged = pd.read_csv('merged.csv')

# Optional: Verify the data loaded correctly
print(df_merged.head())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_merged, x='category', y='purchase_amount', estimator=sum, errorbar=None, palette='viridis')
plt.title('Total Revenue by Category')
plt.ylabel('Total Revenue (USD)')
plt.show()
```

```python
# 2. Is there a relationship between ratings and spending?
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df_merged, x='review_rating', y='purchase_amount', scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Correlation: Review Rating vs. Purchase Amount')
plt.show()
```

```python
# 3. How does shipping type affect purchase volume?
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_merged, x='shipping_type', palette='muted')
plt.title('Frequency of Shipping Types')
plt.show()
```

```python
# 4. Are discounts actually driving higher spending?
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_merged, x='discount_applied', y='purchase_amount', palette='pastel')
plt.title('Impact of Discounts on Purchase Amount')
plt.show()
```

```python
# 5. What are the seasonal sales trends?
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_merged, x='season', y='purchase_amount', errorbar=None, palette='coolwarm')
plt.title('Average Spending by Season')
plt.show()
```

```python
# 6. Which payment methods are most popular in each category?
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_merged, x='category', hue='payment_method')
plt.title('Payment Method Preference by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. How are purchase amounts distributed (Outlier view)?
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_merged, x='purchase_amount', kde=True, color='purple')
plt.title('Distribution of Purchase Amounts')
plt.show()
```

```python
# 8. What is the relationship between size and price?
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_merged[df_merged['category'] == 'Clothing'], x='size', y='purchase_amount', palette='Set2')
plt.title('Price Distribution by Clothing Size')
plt.show()
```

```python
# 9. Which items are purchased most frequently? (Top 10)
```

```python
plt.figure(figsize=(10, 8))
top_items = df_merged['item_purchased'].value_counts().head(10).index
sns.countplot(data=df_merged, y='item_purchased', order=top_items, palette='rocket')
plt.title('Top 10 Most Frequently Purchased Items')
plt.show()
```

```python
# 10. How do different product categories compare in price range?
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_merged, x='category', y='purchase_amount', palette='Set3')
plt.title('Price Range Variability by Category')
plt.xticks(rotation=45)
plt.show()
```
