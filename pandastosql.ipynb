```python
from sqlalchemy import create_engine

# Your MySQL Server details (check these in your MySQL Workbench connection settings)
username = "root"             # Default is usually 'root'
password = "your_password"    # The password you set when installing MySQL
host     = "localhost"        # Usually 'localhost' or '127.0.0.1'
port     = "3306"             # Default MySQL port
database = "retailpulse"

# Build connection string for MySQL
# Format: mysql+pymysql://username:password@host:port/database
connection_str = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"

# Create engine
engine = create_engine(connection_str)

# Test the connection
try:
    with engine.connect() as connection:
        print("✔ Successfully connected to MySQL Workbench!")
except Exception as e:
    print(f"❌ Error connecting to MySQL: {e}")
```

```python
import pandas as pd

# 1. Load the file first (replace 'your_file_name.csv' with the actual file name)
df_merged = pd.read_csv('merged.csv')

# 2. Now run your existing code
table_name = "merged"
df_merged.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"✔ Data successfully loaded into table '{table_name}' in database 'retailpulse'.")
```
