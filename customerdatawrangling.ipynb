```python
# Create a summary DataFrame
null_info = pd.DataFrame({
    'Null Count': df_customer.isnull().sum(),
    'Data Type': df_customer.dtypes
})

# Filter to show only columns that have at least one null value
null_info[null_info['Null Count'] > 0]
```

```python
# Check unique values in columns with nulls to see if they look like placeholders
print("Unique values in Subscription Status:", df_customer['Subscription Status'].unique())
```

```python
import numpy as np

# Convert 'na' and 'not available' to actual NaN
df_customer['Subscription Status'] = df_customer['Subscription Status'].replace(['na', 'not available'], np.nan)
```

```python
# Check unique values in columns with nulls to see if they look like placeholders
print("Unique values in Subscription Status:", df_customer['Subscription Status'].unique())
```

```python
# Fill 'Subscription Status' with the most frequent value (Mode)
mode_sub = df_customer['Subscription Status'].mode()[0]
df_customer['Subscription Status'] = df_customer['Subscription Status'].fillna(mode_sub)
```

```python
print("Unique values in Subscription Status:", df_customer['Subscription Status'].unique())
```

```python
print("Unique values in Previous Purchases:", df_customer['Previous Purchases'].unique())
```

```python
# Filling Previous Purchases with Zero(0)
df_customer['Previous Purchases']=df_customer['Previous Purchases'].fillna(0)
```

```python
print("Unique values in Previous Purchases:", df_customer['Previous Purchases'].unique())
```

```python
print("Unique values in Frequency of Purchases:", df_customer['Frequency of Purchases'].unique())
```

```python
# Fill with a new category
df_customer['Frequency of Purchases'] = df_customer['Frequency of Purchases'].fillna('Unknown')
```

```python
# Create a summary DataFrame
null_info = pd.DataFrame({
    'Null Count': df_customer.isnull().sum(),
    'Data Type': df_customer.dtypes
})

# Filter to show only columns that have at least one null value
null_info[null_info['Null Count'] > 0]
```
